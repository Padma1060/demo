```python
#palindrome or not
a=input('enter the string')
if a==a[::-1]:
    print("the string is palindrome")
else:
    print("the string is not palindrome")
```

```python
# 1 to 20
# 15->jackpot number
# 10
# choose the higher number
#19
#choose the lower number
#15
#your attempt is 3 times
import random
jackpot=random.randint(1,20)
user_num=int(input('enter the number'))
c=1
while jackpot!=user_num
   if user_num<jackpot:
      print('choose higher num')
   else:
       print('choose lower num')
   user_num=int(input('enter the number'))
   c+=1
else:
    print('you won the game and your jackpot num is',user_num)
    print('your attempt',c)
```
